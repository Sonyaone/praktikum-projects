#### Заполним пропущенные значения в living_area и kitchen_area.

Заполним эти поля используя квартили и проверим заполнение:

```python
quantiles = [0.25, 0.5, 0.75]
def fill_nan_area(data):
# Функция итерирует список с квантилямя и при условии, что значение в столбце total_area меньше следующего квантиля
    # этого столбца, заполняет пропущенные значения данного квантиля столбцов kitchen_area и Living_area
    for x in quantiles:
        total_area = real_estate_data['total_area']
        real_estate_data.loc[total_area <= total_area.quantile(x+0.25), 'living_area'] = real_estate_data.loc[
            total_area <= total_area.quantile(x+0.25), 'living_area'].fillna(real_estate_data['living_area'].quantile(x))
        real_estate_data.loc[total_area <= total_area.quantile(x+0.25), 'kitchen_area'] = real_estate_data.loc[
            total_area <= total_area.quantile(x+0.25), 'kitchen_area'].fillna(real_estate_data['kitchen_area'].quantile(x))
fill_nan_area(real_estate_data)


real_estate_data['living_area'].isna().sum()
real_estate_data['kitchen_area'].isna().sum()
```

```python
def cityCenter_group(city):
    """
    Разделяет квартиры на три категории по удаленности от цента.
    """
    if  399 <= city <= 10000:
        return 'до 10 км'
    if  10001 <= city <= 20000:
        return 'до 20 км'
    if  20001 <= city <= 30000:
        return 'до 30 км'
    if  30001 <= city:
        return 'более 30 км'
    
    
sample_estate_data['center_distance_group'] = sample_estate_data['cityCenters_nearest'].apply(cityCenter_group)
sample_estate_data.head(10)
```
